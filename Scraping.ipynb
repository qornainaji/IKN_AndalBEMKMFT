Sekarang kita akan membuat perintah untuk scraping

```python
c = twint.Config()

#Coba pilih 2 topik bebas, gunakan OR untuk menggabungkan. Jangan lupa pakai string.
# c.Search = "ibu kota negara OR ibu kota negara baru OR Ibu Kota Negara Baru OR ibu kota nusantara"
c.Search = "ibu kota negara"
# c.Search = "ibu kota negara baru" 
# c.Search = "Ibu Kota Negara Baru" 
# c.Search = "ibu kota nusantara"
c.Location = True

#Di sini, kamu boleh memfilter dengan kriteria yang kamu inginkan agar tweet tidak terlalu banyak (Misal rentang tanggal tweet)
c.Lang = "id"
c.Since = "2021-01-01"
c.Until = "2022-06-21"
c.Filter_retweets = True



#Pastikan sudah dalam bentuk csv
c.Store_csv = True

c.Output = "Location.csv"
```

Petunjuk terkait scraping: 
- Konfigurasi twint : https://github.com/twintproject/twint/wiki/Configuration
- Query pencarian Twitter : https://github.com/igorbrigadir/twitter-advanced-search

#### Twint, do your magic!!!

```python
twint.run.Search(c)
```
